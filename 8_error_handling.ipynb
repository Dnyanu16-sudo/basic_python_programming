```python
# 1. try-except
# try contains code that might cause an error. If an error happens, except handles it.
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# 2. Multiple except blocks
# Different errors can be handled in different ways using multiple except blocks.
```

```python
try:
    num = int("abc")
    result = 10 / num
except ValueError:
    print("Conversion failed")
except ZeroDivisionError:
    print("Division by zero error")
```

```python
# 3. Multiple exceptions in one except
# You can handle more than one error type in a single block.
```

```python
try:
    x = int("abc")
except (ValueError, TypeError):
    print("Value or Type error occurred")
```

```python
# 4. Generic Exception
# Catches any kind of error when you don’t specify the exact type.
```

```python
try:
    x = int("abc")
except Exception as e:
    print("Error occurred:", e)
```

```python
# 5. else block
# Runs only when no error occurs in the try block.
```

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("Success:", x)
```

```python
# 6. finally block
# Runs every time, whether an error happens or not. Often used for cleanup.
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("This always runs")
```

```python
# 7. Full structure
# All parts used together for complete handling.
```

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("Result:", x)
finally:
    print("Execution complete")
```

```python
# 8. raise keyword
# Used to manually create and throw an error.
```

```python
def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18+")
    return "Allowed"

try:
    check_age(15)
except ValueError as e:
    print(e)
```

```python
# 9. Custom Exception
# You can create your own error type for specific situations.
```

```python
class CustomError(Exception):
    pass

def test():
    raise CustomError("This is a custom error")

try:
    test()
except CustomError as e:
    print(e)
```

```python
# 10. assert statement
# Checks a condition; if false, it raises an error.
```

```python
x = -5
try:
    assert x > 0, "x must be positive"
except AssertionError as e:
    print(e)
```

```python
# 11. Nested try-except
# A try-except block placed inside another.
```

```python
try:
    try:
        x = int("abc")
    except ValueError:
        print("Inner error")
        raise
except Exception:
    print("Outer handler caught it")
```

```python
# 12. File handling with exceptions
# Used to handle errors while working with files.
```

```python
try:
    f = open("file.txt", "r")
    data = f.read()
except FileNotFoundError:
    print("File not found")
finally:
    try:
        f.close()
    except:
        pass
```
