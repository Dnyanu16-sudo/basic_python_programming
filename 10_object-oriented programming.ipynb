```python
# Method overriding 
```

```python
# Method overriding replaces a parent method in the child class
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")


class Car(Vehicle):
    def start(self):
        print("Car is starting...")
```

```python
# Using Super
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")


class Car(Vehicle):
    def start(self):
        super().start()  # Call parent method
        print("Car is starting...")
```

```python
car = Car()
```

```python
car.start()
```

```python
# Abstract classes
```

```python
# ABC is a class used to force child classes to implement certain methods
# @abstractmethod marks methods that child classes must provide
```

```python
# ABC = a rule book
# @abstractmethod = compulsory method
```

```python
# An abstract class is a class that cannot be instantiated and is used as a blueprint for other classes
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):

    def area(self):
        return 78.5
```

```python
Circle().area()
```

```python
from abc import ABC, abstractmethod


class Calculator(ABC):

    @abstractmethod
    def calculate(self):
        pass


class Addition(Calculator):

    def __init__(self, a, b):
        self.a = a
        self.b = b

    def calculate(self):
        return self.a + self.b
```

```python
add = Addition(10, 20)
```

```python
print(add.calculate())
```

```python
# Decorator
```

```python
# A decorator is a function that adds extra functionality to another function without modifying the original function
```

```python
def decorator(func):

    def wrapper():
        print("Before function")
        func()
        print("After function")

    return wrapper


@decorator
def greet():
    print("Hello")
```

```python
greet()
```

```python
# @property is a decorator that allows a method to be accessed like an attribute (variable)
```

```python
# @property is a decorator that converts a method into a read-only attribute, allowing it to be accessed without parentheses
```

```python
# Without @property
```

```python
class Student:
    def get_name(self):
        return "Rahul"

s = Student()

print(s.get_name())
```

```python
# With @property
```

```python
class Student:

    @property
    def name(self):
        return "Rahul"

s = Student()

print(s.name)
```
